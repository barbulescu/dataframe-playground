```kotlin
%useLatestDescriptors
%use dataframe

val df = DataFrame.readCsv(
    "https://raw.githubusercontent.com/Kotlin/dataframe/master/data/jetbrains_repositories.csv",
)

df.describe()
```

```kotlin
df
```

```kotlin
val columns = df.select { full_name and stargazers_count and topics}
columns
```

```kotlin
%use kandy

df.plot {
    bars {
        x(full_name)
        y(stargazers_count)
    }

    layout.title = "Top 10 JetBrains repositories by stars count"
}
```

```kotlin
val dfFiltered = columns.filter { stargazers_count >= 1000 }
dfFiltered
```
